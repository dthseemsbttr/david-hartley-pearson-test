```python
import os
import pandas as pd
```

```python
files = os.listdir('.\Котировки')
files
```

```python
def get_year(x):
    return str(x)[:4]

df = pd.DataFrame(columns=['Ticker', '2015', '2016', '2017', '2018', '2019', '2020', '2021'])
for file_name in files:
    temp_df = pd.read_csv(f'.\Котировки\{file_name}')
    output = [temp_df['<TICKER>'][0]]
    for y in ['2015', '2016', '2017', '2018', '2019', '2020', '2021']:
        output.append(temp_df[temp_df['<DATE>'].apply(get_year) == y]['<DATE>'].count())
    df.loc[len(df)] = output
df = df.set_index(['Ticker'])
df
```

```python
df.to_csv('.\Результаты\Результат торгов по дням.csv', sep=';')
```
