```python
import matplotlib.pyplot as plt
import pandas as pd
```

## NVIDIA

```python
df_NVDA = pd.read_csv(r'.\Котировки\US1.NVDA_150101_220101.csv')
df_NVDA['<DATE>'] = pd.to_datetime(df_NVDA['<DATE>'], format="%Y%m%d")
df_NVDA = df_NVDA.set_index(['<DATE>'])
df_NVDA.head()
```

```python
plt.figure(1, (15, 10))
df_NVDA['<CLOSE>'].plot()
plt.title('График цен акций компании NVIDIA')
plt.xlabel('Года')
plt.ylabel('Цена, дол.')
plt.grid()
plt.show()
```
