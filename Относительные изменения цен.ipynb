```python
import os
import pandas as pd
```

```python
files = os.listdir('.\Котировки')
del files[8]
files
```

```python
adf = pd.read_csv(f'.\Котировки\{files[0]}')
adf.head()
```

```python
def get_year(x):
    return str(x)[:4]

def change_sep(x):
    return str(round(x, 4)).replace('.', ',')

df_high = pd.DataFrame(columns=['Ticker', '2015', '2016', '2017', '2018', '2019', '2020', '2021'])
df_low = pd.DataFrame(columns=['Ticker', '2015', '2016', '2017', '2018', '2019', '2020', '2021'])

for file_name in files:
    temp_df = pd.read_csv(f'.\Котировки\{file_name}')
    temp_df['<CHANGES>'] = temp_df['<CLOSE>'].pct_change().round(4) * 100
    temp_df['<YEAR>'] = temp_df['<DATE>'].apply(get_year)
    high = [temp_df['<TICKER>'][0]] + temp_df.groupby(['<YEAR>'])['<CHANGES>'].max().apply(change_sep).to_list()
    low = [temp_df['<TICKER>'][0]] + temp_df.groupby(['<YEAR>'])['<CHANGES>'].min().apply(change_sep).to_list()
    df_high.loc[len(df_high)] = high
    df_low.loc[len(df_low)] = low
df_high = df_high.set_index(['Ticker'])
df_low = df_low.set_index(['Ticker'])
```
